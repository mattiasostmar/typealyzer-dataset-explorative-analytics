# ESTJ vs INTJ and ISTJ vs ENTJ (opposite types)
Source: Gerald P. Macdaid CAPT Data Bank 1997 
[Center for Applications of Psychological Type Inc.](http://www.myersbriggs.org/my-mbti-personality-type/mbti-basics/type-tables.htm?bhcp=1)

```python
Image("../reports/figures/myers-briggs-matrix.gif")
```

### ESTJ vs INTJ

```python
estj_mean = df[df["actual"] == "ESTJ"].mean()
estj_mean.name = "ESTJ"
intj_mean = df[df["actual"] == "INTJ"].mean()
intj_mean.name = "INTJ"
```

```python
overview = pd.concat([estj_mean.describe(),intj_mean.describe()], axis = 1)
overview
```

```python
estj_intj_df = pd.concat([estj_mean,intj_mean], axis=1)
estj_intj_df.plot()
```

### Preliminary conclusion: 
There are **NO** lingustic patterns differentiating self-reported introversion and extroversion respectively in the Typealyzer survey. 

There are slight variations in LIWC category percentages though, which hints at that the LIWC word counting works correctly.

### ISTJ vs ENTJ

```python
istj_mean = df[df["actual"] == "ISTJ"].mean() 
entj_mean = df[df["actual"] == "ENTJ"].mean()
```

```python
istj_entj_df = pd.concat([istj_mean, entj_mean], axis=1)
istj_entj_df.plot()
```

### Preliminary conclusion: 
There are **NO** lingustic patterns differentiating self-reported introversion and extroversion respectively in the Typealyzer survey. 

There are slight variations in LIWC category percentages though, which hints at that the LIWC word counting works correctly.
