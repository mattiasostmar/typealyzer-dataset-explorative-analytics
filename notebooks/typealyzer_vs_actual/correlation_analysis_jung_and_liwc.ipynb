```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_pickle("../pickles/dataframe_survey_2018-01-23_jung-liwc-dt-enriched.pickle")
```

```python
liwc = ['adverb', 'verb', 'bio', 'sexual', 'they', 'social', 'ppron', 'article', 'sad', 'number', 'swear', 'pronoun', 'affect', 'friend', 'incl', 'work', 'time', 'funct', 'auxverb', 'cause', 'filler', 'anger', 'home', 'family', 'money', 'quant', 'i', 'achieve', 'nonfl', 'tentat', 'see', 'future', 'insight', 'excl', 'we', 'inhib', 'past', 'negemo', 'posemo', 'ipron', 'cogmech', 'you', 'death', 'leisure', 'assent', 'percept', 'ingest', 'negate', 'discrep', 'conj', 'body', 'feel', 'hear', 'present', 'space', 'relativ', 'relig', 'shehe', 'motion', 'humans', 'certain', 'anx', 'health', 'preps']
jung = jung = ["att","func","funcatt","actual"]
corr_df = df[liwc + jung]
```

```python
# Correlates 
corr = corr_df.corr()

ax = plt.axes()
sns.heatmap(corr, 
        ax=ax,
        xticklabels=corr.columns,
        yticklabels=corr.columns,
        cmap="BuPu")
sns.set()
sns.set_context("poster")
ax.set_title('Pearson correlation Jung vs LIWC in Typealyzer survey (n=27959)')
plt.show()
```
